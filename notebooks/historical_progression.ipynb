## The maintained lesson

The canonical module collects several reusable approaches: direct addition, built-in `sum`, `reduce`, and `math.fsum`. Importing it keeps this notebook aligned with the maintained source instead of copying implementation code.


````python
from demos.summing_methods import add_operator, add_sum, sum_builtin, sum_fsum, sum_reduce

numbers = [1.5, 2.25, -0.75]
{
    "a + b": add_operator(7, 5),
    "sum((a, b))": add_sum(7, 5),
    "sum(numbers)": sum_builtin(numbers),
    "reduce(add, numbers)": sum_reduce(numbers),
    "math.fsum(numbers)": sum_fsum(numbers),
 }
````

For the interactive lesson, run `python -m demos.summing_methods`. For a scriptable one-shot sum, run `python -m demos.summing_methods --numbers 3 5`; add `--float` for finite floating-point values.
